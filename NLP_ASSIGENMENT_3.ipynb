1. Create a Doc object from the file peterrabbit.txt


```python
import spacy
nlp = spacy.load("en_core_web_sm")
text = open("/content/peterrabbit.txt", encoding="utf-8").read()
doc = nlp(text)
```

2. For every token in the third sentence, print the token text, the POS tag, the fine-grained TAG tag, and the description of the fine-grained tag.

```python
import spacy

sentences = list(doc.sents)
third_sentence = sentences[2]

# Print details for each token
for token in third_sentence:
    print(f"{token.text:<15} {token.pos_:<10} {token.tag_:<10} {spacy.explain(token.tag_)}")
```

3. Provide a frequency list of POS tags from the entire document

```python
import spacy
from collections import Counter
# Count how often each POS tag appears
pos_counts = Counter([token.pos_ for token in doc])

# Print the results
for pos, count in pos_counts.items():
    print(pos, ":", count)
```

4. CHALLENGE: What percentage of tokens are nouns?

```python
# Count total tokens and nouns
total_tokens = len([token for token in doc if not token.is_punct])
noun_tokens = len([token for token in doc if token.pos_ == "NOUN"])

# Calculate percentage
percentage = (noun_tokens / total_tokens) * 100

print(f"Percentage of tokens that are nouns: {percentage:.2f}%")
```

5. Display the Dependency Parse for the third sentence.

```python
third_sentence = list(doc.sents)[2]

for token in third_sentence:
    print(f"{token.text:<15} {token.dep_:<10} {token.head.text:<15} ({token.head.pos_})")
```

6. Show the first two named entities from Beatrix Potter's The Tale of Peter Rabbit


```python
for ent in list(doc.ents)[:2]:
    print(ent.text, "-", ent.label_)
```

7. How many sentences are contained in The Tale of Peter Rabbit?

```python
num_sentences = len(list(doc.sents))

print("Number of sentences in The Tale of Peter Rabbit:", num_sentences)
```

8. CHALLENGE: How many sentences contain named entities?

```python
sentences_with_ents = [sent for sent in doc.sents if any(token.ent_type_ for token in sent)]
count = len(sentences_with_ents)

print("Number of sentences containing named entities:", count)
```

9. Display the named entity visualization for list_of_sents[0] from the previous problem

```python
from spacy import displacy

# Find sentences that contain at least one named entity
list_of_sents = [sent for sent in doc.sents if any(token.ent_type_ for token in sent)]

# Display the named entity visualization for the first one
displacy.render(list_of_sents[0], style="ent", jupyter=True)
```
