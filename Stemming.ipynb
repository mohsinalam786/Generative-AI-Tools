```python
# Import tghe toolkit and the full Porter Stemmer Library
import nltk
from nltk.stem.porter import PorterStemmer
```

### Porter stemmer omne of the most common -  and effective-stemming tools is Porter's Algorithm developed by Martin

```python
p_stemmer = PorterStemmer()
```

```python
words= ["run", "runner", "ran", "runs", "easily", "fairly"]
```

```python
for word in words:
  print(word+"-->"+p_stemmer.stem(word))
```

## Snowball Stemmer

```python
from nltk.stem.snowball import SnowballStemmer

# The SnowBall Stemmer requires that you pass a launguage parameter
s_stemmer = SnowballStemmer(language="english")
```

```python
words = ["run", "runner","running", "ran", "runs", "easily", "fairly"]
# words = ["Generous", "generation","generously", "generate"]
```

```python
for word in words:
  print(word+"-->"+s_stemmer.stem(word))
```
