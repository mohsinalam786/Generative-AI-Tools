# Stop Words
### Words like "a" and "the" appear so frequently that they don't require tagging as thoroughly as nouns, verbs, modifiers. We call these *stop words*, and they can be filtered from the text to be processed. Spacy hold a built-in list of some 305 English stop words.

```python
# Perform standard inputs
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Print the set of spacy's default stop words(remember)
print(nlp.Defaults.stop_words)
```

```python
print(len(nlp.Defaults.stop_words))
```

To see if a word is stop word

```python
nlp.vocab["myself"].is_stop
```

```python
nlp.vocab["mystery"].is_stop
```

```python
nlp.vocab["beyond"].is_stop
```

 To add a stop word

```python
nlp.vocab["www"].is_stop
```

```python
# Add th word to the set of stop words. Use lowercase!
nlp.Defaults.stop_words.add("www")

# Set the stop word tag on the lexeme
nlp.vocab["www"].is_stop
```

```python
print(len(nlp.Defaults.stop_words))
```

```python
nlp.vocab["www"].is_stop
```

Removing a stop word

```python
nlp.vocab["last"].is_stop
```

```python
# Remove the word from the set of stop words
nlp.Defaults.stop_words.remove("last")
```

```python
#Remove the stop_word tag from the lexeme
nlp.vocab["last"].is_stop= False
```

```python
print(len(nlp.Defaults.stop_words))
```
