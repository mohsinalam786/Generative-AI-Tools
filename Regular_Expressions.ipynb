```python
#Finding pattern
```

```python
text="My telephone number is 408-555-1234"
```

```python
phone=re.search(r'\d\d\d-\d\d\d-\d\d\d\d',text)
```

```python
phone.group()
```

```python
phone_pattern=re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
results = re.search(phone_pattern,text)
```

```python
results.group()
```

```python
results.group(1)
```

```python
results.group(2)
```

```python
results.group(3)
```
