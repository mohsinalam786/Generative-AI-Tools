```python
# Perform standard imports
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Import the Matcher library
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

Creating Pattterns
###### In literature, the phrase "solar power"

```python
pattern1 = [{"LOWER":"solarpower"}]
pattern2 = [{"LOWER": "solar"}, {"LOWER":"power"}]
pattern3 = [{"LOWER": "solar"}, {"IS_PUNCT": True}, {"LOWER": "power"}]
matcher.add("SolarPower", [pattern1, pattern2, pattern3])
```

```python
doc = nlp(u"The Solar Power industry continues to grow as demand\ for solarpower increases. Solar-power cars are getting populariry.")
```

```python
found_matches =matcher(doc)
print(found_matches)
```

```python
for match_id, start, end in found_matches:
  string_id = nlp.vocab.strings["match_id"]
  span = doc[start:end]
  print(match_id, string_id, start, end, span.text)
```

```python
# Redefine the patterns
pattern1 = [{"LOWER": "solarpower"}]
pattern2 = [{"LOWER": "solar"}, {"IS_PUNCT": True, "OP":"*"}, {"LOWER":"power"}]
```

```python
# Add the new set of patterns to the "SolarPower" matcher:
matcher.add("SolarPower", [pattern1, pattern2])
```

```python
found_matches =  matcher(doc)
print(found_matches)
```

```python
pattern1 = [{"LOWER": "solarpower"}]
pattern2 = [{"LOWER": "solar"}, {"IS_PUNCT":True, "OP":"*"}, {"LEMMA":"power"}]
matcher.remove("SolarPower")
matcher.add("SolarPower", [pattern1, pattern2])
```

```python
doc2 = nlp(u"Solar-powered energy runs solar-powered cars.")
```

```python
matches_found = matcher(doc2)
print(matches_found)
```

## Phrase Matcher

```python
#Perform standard inputes, reset nlp
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
# Import the PhraseMatcher Library
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab)
```

```python
with open("reaganomics.txt", "r", encoding='latin-1') as f:
  doc3 = nlp(f.read())
```

```python
# First, create a list of match phrases:
phrase_list = ["voodoo economics", "supply-side economics" ,"trickle-down economics", "free-market economics"]
```

```python
# Next, convert each phrase to a Doc object
phrase_patterns  = [nlp(text) for text in phrase_list]
```

```python
# Pass each Doc object object into matcher(note the use of the asterisk!)
matcher.add("VoodooEconomics", None, *phrase_patterns)
```

```python
# Build a list of matches
matches = matcher(doc3)
```
