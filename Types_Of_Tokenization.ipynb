Word based tokenization

```python
s  = "I will take this course."
```

```python
s.split()
```

Character based tokenization

```python
List = []
for char in s:
  List.append(char)
print(List)
```

Removing white spaces

```python
text = "".join(s.split())
print(text)
```

```python
List2 =[]
for char in text:
  List2.append(char)
print(List2)
```

Subword Tokenization

```python
string = "I am learning Tokenization"
```

```python
newString = string.split()
```

```python
newString
```

```python
List3 = []
for word in newString:
  if word[-3:] == "ing":
    List3.append(word.partition("ing"))
  elif word[-7:] == "ization":
    List3.append(word.partition("ization"))
  else:
    List3.append(word)
print(List3)
```

```python
List4 = []
for element in List3:
  if type(element) is tuple:
    List4.append(element[0])
    List4.append(element[1])
  else:
    List4.append(element)
print(List4)
```
