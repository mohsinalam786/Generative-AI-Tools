Import Libraries

```python
import nltk
from nltk import word_tokenize
```

**Natural Language Toolkit(NLTK)**
##NLTK is an open source library. It provides many functionalities and include pre-trained model

  NLTK Model Language

```python
nltk.download("punkt_tab") #It is pre-trained model for tokenization
```

```python
sentence = "Today is Friday. Next week, we're coming from U.S.!"
```

```python
for token in word_tokenize(sentence):
  print(token)
```

```python
sentence.split()
```

```python
sentence2 = "We're here to help you! ask the question, email support@abcnet.ax or visit us at https://www.abcnet.ax!"
```

```python
for token in word_tokenize(sentence2):
  print(token)
```
